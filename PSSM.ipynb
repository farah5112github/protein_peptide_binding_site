```
# find out the updated seq_code :
updated_seq_code = []

def seperate_the_data_sets_into_lists(file_path):
    f = open(file_path)
    file = f.readlines()

    for i in list(    range(  0,len(file) )     ):
        s = file[i]
        code = s[:-1]
        updated_seq_code.append(code)
    f.close()

# call the above function to find code and sequnce : 
file_path = './gdrive/My Drive/Fall 2020/project_codes/updated_seq_code.txt'
seperate_the_data_sets_into_lists(file_path)

print('The updated codes : ',len(updated_seq_code));
print(updated_seq_code)
```

```
# find out the sequences from file :
np.set_printoptions(threshold=np.inf)
total_input = [] 
seq_code = []
count = 0 ;

def seperate_the_data_sets_into_lists_1(file_path):
    f = open(file_path)
    file = f.readlines()

    
    for i in list(    range(  0,len(file),3  )     ):
        s = file[i]
        code = s[:-1]
        if code in updated_seq_code and code not in seq_code:
           
           seq_code.append(code)
           input_index = i + 1 
           s = file[input_index]
           seq = s[:-1]
           total_input.append(seq)
        
    f.close()

# call the above function to find code and sequnce : 
file_path = './gdrive/My Drive/Fall 2020/project_codes/sample_FASTA_file.txt'
seperate_the_data_sets_into_lists_1(file_path)

print(len(seq_code))
print(len(total_input))
```

```
# PSSM individual probability : 20 : ws = 9 ;
import math
PSSM_total_9 = np.zeros((1,20),dtype=float)

def find_individual_PSSM_files(file_path):
    
    f = open(file_path,'r')
    file = f.readlines()
    file = file[3:-6]
    PSSM_indi_9 = np.zeros( ( len(file),20  ),dtype=float)

    for i in range( 0,len(file) ):
        line = file[i]
        line = line.split(' ')
        s = []        
        
        for ele in line: 
            if ele.strip(): 
               s.append(ele)
        s = s[2:-22] 
        
        for j in range(0,20):
            PSSM_indi_9[i,j] = float(s[j])        
        
    dummy_prev_four = np.concatenate(  (  PSSM_indi_9[4,:],PSSM_indi_9[3,:],PSSM_indi_9[2,:],PSSM_indi_9[1,:] ), axis=0  )
    dummy_prev_four = np.reshape(dummy_prev_four,(4,20))
    
    dummy_last_four = np.concatenate(  (  PSSM_indi_9[len(file)-2,:],PSSM_indi_9[len(file)-3,:],PSSM_indi_9[len(file)-4,:],PSSM_indi_9[len(file)-5,:] ), axis=0  )
    dummy_last_four = np.reshape(dummy_last_four,(4,20))

    PSSM_indi_9 = np.concatenate( (dummy_prev_four,PSSM_indi_9,dummy_last_four),axis=0)
    
    
    r,c = PSSM_indi_9.shape

    for row in list(range(0,r)):
        for col in list(range(0,c)):
            PSSM_indi_9[row,col] = math.exp(   PSSM_indi_9[row,col]  ) / (   1 + math.exp(   PSSM_indi_9[row,col]  ) )

    global PSSM_total_9 
    PSSM_total_9 = np.concatenate( (PSSM_total_9,PSSM_indi_9),axis=0)    
        
    

# evoluationary info :
for i in list(  range( 0,len(seq_code) )  ):
    file_path = str('./gdrive/My Drive/Fall 2020/project_codes/pssm/' + seq_code[i] + '.pssm')
    find_individual_PSSM_files(file_path)
    #print('-------------------------------------------------------------------------------------------------')

PSSM_total_9 = np.delete(PSSM_total_9, 0, 0)
print('PSSM values shape : window - 9 : ' , PSSM_total_9.shape)

# Info_entropy : 1 , ws = 9

import math
Info_entropy_total_9 = np.zeros((1,1),dtype=float)

def find_individual_PSSM_files(file_path):
    
    f = open(file_path,'r')
    file = f.readlines()
    file = file[3:-6]
    PSSM_indi_9 = np.zeros( ( len(file),20  ),dtype=float)
    

    for i in range( 0,len(file) ):
        line = file[i]
        line = line.split(' ')
        s = []        
        
        for ele in line: 
            if ele.strip(): 
               s.append(ele)
        s = s[2:-22] 
        
        for j in range(0,20):
            PSSM_indi_9[i,j] = float(s[j])        
        
    dummy_prev_four = np.concatenate(  (  PSSM_indi_9[4,:],PSSM_indi_9[3,:],PSSM_indi_9[2,:],PSSM_indi_9[1,:] ), axis=0  )
    dummy_prev_four = np.reshape(dummy_prev_four,(4,20))
    
    dummy_last_four = np.concatenate(  (  PSSM_indi_9[len(file)-2,:],PSSM_indi_9[len(file)-3,:],PSSM_indi_9[len(file)-4,:],PSSM_indi_9[len(file)-5,:] ), axis=0  )
    dummy_last_four = np.reshape(dummy_last_four,(4,20))

    PSSM_indi_9 = np.concatenate( (dummy_prev_four,PSSM_indi_9,dummy_last_four),axis=0)
    
    
    r,c = PSSM_indi_9.shape

    for row in list(range(0,r)):
        for col in list(range(0,c)):
            PSSM_indi_9[row,col] = math.exp(   PSSM_indi_9[row,col]  ) / (   1 + math.exp(   PSSM_indi_9[row,col]  ) )

    #PSSM_info_entropy_indi : 
    Info_entropy_indi_9 = np.zeros( (r,1),dtype=float)
    for row in list(range(0,r)):
        sum = 0
        for col in list(range(0,c)):
            sum = sum +  (  PSSM_indi_9[row,col] * np.log( PSSM_indi_9[row,col] ) ) * ( -1 )
        Info_entropy_indi_9[row,0] = sum 

    global Info_entropy_total_9 
    Info_entropy_total_9 = np.concatenate( (Info_entropy_total_9,Info_entropy_indi_9),axis=0)    

# evoluationary info :
for i in list(  range( 0,len(seq_code) )  ):
    file_path = str('./gdrive/My Drive/Fall 2020/project_codes/pssm/' + seq_code[i] + '.pssm')
    find_individual_PSSM_files(file_path)
    #print('-------------------------------------------------------------------------------------------------')

Info_entropy_total_9 = np.delete(Info_entropy_total_9, 0, 0)
print('Info entropy values shape : window - 9 : ' , Info_entropy_total_9.shape)


```
